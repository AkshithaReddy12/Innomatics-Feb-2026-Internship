```python
# Task 1: User Login Check

# Predefined login credentials
username = "admin"
password = "1234"

# Taking input from user
entered_user = input("Enter username: ")
entered_pass = input("Enter password: ")

# Check if both username and password match
if entered_user == username and entered_pass == password:
    print("Login Successful")   # If credentials are correct
else:
    print("Invalid Credentials")  # If credentials are wrong
```

```python
# Task 2: Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

# Counters for pass and fail students
pass_count = 0
fail_count = 0

# Loop through each mark
for m in marks:
    if m >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Display results
print("Total Passed Students:", pass_count)
print("Total Failed Students:", fail_count)
```

```python
# Task 3: Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []   # Empty list to store cleaned names

# Clean each name
for name in names:
    cleaned_names.append(name.strip().lower())

# Display cleaned list
print("Cleaned Names List:", cleaned_names)
```

```python
# Task 4: Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

# Loop through each message
for msg in messages:
    print("Message:", msg)
    print("Length:", len(msg))   # len() gives number of characters

    # Flag message if length is greater than 10
    if len(msg) > 10:
        print("Flag: Long message")

    print()  # Blank line for better readability
```

```python
# Task 5: Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0   # Counter for ERROR entries

# Loop through logs
for log in logs:
    if log == "ERROR":   # Check if log is ERROR
        error_count += 1

# Display total ERROR count
print("Total ERROR entries:", error_count)
```
